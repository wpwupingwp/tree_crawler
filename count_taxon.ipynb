```python
with open('lineage_query.out', 'r') as _:
    data = _.readlines()
    data = [i.rstrip() for i in data]
    data = [i.split('\t') for i in data]
print(data[:10])
```

```python
result = dict(Other=0, Archaea=0, Bacteria=0, Fungi=0, Metazoa=0, Viridiplantae=0,Viruses=0)
print(result)
```

```python
other = list()
for i in data:
    found = False
    for j in reversed(i):
        if ',' in j:
            j == j.split(',')[0]
        if j in result:
            result[j] += 1
            found = True
            break
        elif j.endswith('phyceae'):
            result['Viridiplantae'] +=1
            found = True
            pass
        elif j in {'Influenza'}:
            result['Viruses'] += 1
            found = True
        else:
            continue
    if not found: 
        result['Other'] += 1
        other.append(i)
```
